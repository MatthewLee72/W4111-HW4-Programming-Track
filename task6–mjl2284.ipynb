## Health Check

```python
response = requests.get(f'{BASE_URL}/health')
print(f'Status Code: {response.status_code}')
print(json.dumps(response.json()))
```

## 1. Customer API

```python
# Get all customers
response = requests.get(f'{BASE_URL}/customers')
print(f'Status Code: {response.status_code}')
print(json.dumps(response.json()))
```

```python
# Create new customer
new_customer = {
    'customerName': 'Matt',
    'contactLastName': 'Lee',
    'contactFirstName': 'Matt',
    'phone': '911',
    'addressLine1': '123 Hacker Highway',
    'city': 'Morningside Heights',
    'country': 'Upper West Side'
}
response = requests.post(f'{BASE_URL}/customers', json=new_customer)
print(f'Status Code: {response.status_code}')
new_id = response.json()
print('New User ID:', new_id)
```

```python
# Retrieve newly created customer
response = requests.get(f'{BASE_URL}/customers/{new_id}')
print(f'Status Code: {response.status_code}')
print(json.dumps(response.json()))
```

```python
# Update the customer
updated_customer = new_customer.copy()
updated_customer['phone'] = '119'
response = requests.put(f'{BASE_URL}/customers/{new_id}', json=updated_customer)
print(f'Status Code: {response.status_code}')
print(response.json())
```

```python
# Delete the customer
response = requests.delete(f'{BASE_URL}/customers/{new_id}')
print(f'Status Code: {response.status_code}')
print(response.json())
```

```python
# Show 404 Error for Nonexistent Customer
response = requests.get(f'{BASE_URL}/customers/9999999')
print(f'Status Code: {response.status_code}')
print(json.dumps(response.json()))
```

## 2. Orders API

```python
# Get all orders
response = requests.get(f'{BASE_URL}/orders')
print(f'Status Code: {response.status_code}')
print(json.dumps(response.json()))
```

```python
# Create new order for customer 112
new_order = {
    "orderDate": "2026-05-04",
    "requiredDate": "2026-05-10",
    "status": "In Process",
    "customerNumber": 112
}
response = requests.post(f"{BASE_URL}/orders", json=new_order)
print(f"Status Code: {response.status_code}")
new_order_id = response.json()
print("New Order ID:", new_order_id)
```

```python
# Get our new order
response = requests.get(f"{BASE_URL}/orders/{new_order_id}")
print(f"Status Code: {response.status_code}")
print("Response:", response.json())
```
